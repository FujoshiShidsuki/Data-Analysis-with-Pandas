```python
import pandas as pd

# Load the dataset
data = pd.read_csv('Sample.csv')

# Display the first few rows of the dataset
data.head()
```

```python
#DATA EXPLORATION


# Filter data for a specific region
north_region_data = data[data['Region'] == 'North']
print("Data for North region:")
print(north_region_data)

# Sort data by Quantity in descending order
sorted_data = data.sort_values(by='Quantity', ascending=False)
print("\nData sorted by Quantity (descending):")
print(sorted_data)

# Group data by Product and calculate the total Quantity sold for each product
grouped_data = data.groupby('Product')['Quantity'].sum().reset_index()
print("\nTotal Quantity sold for each product:")
print(grouped_data)
```

```python
# Calculate mean, median, and standard deviation for Quantity

mean_quantity = data['Quantity'].mean()
median_quantity = data['Quantity'].median()
std_quantity = data['Quantity'].std()

print("\nSummary Statistics for Quantity:")
print(f"Mean: {mean_quantity}")
print(f"Median: {median_quantity}")
print(f"Standard Deviation: {std_quantity}")
```

```python
#DATA VISUALISATION

import matplotlib.pyplot as plt
import seaborn as sns

# Plot distribution of Quantity
plt.figure(figsize=(10, 6))
sns.histplot(data['Quantity'], kde=True)
plt.title('Distribution of Quantity Sold')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()

# Plot relationship between Quantity and Price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Quantity', y='Price', hue='Product')
plt.title('Quantity vs Price')
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.show()
```
